```python
%%manim makeText -l

class makeText(Scene):
    def construct(self):
        #######Code#######
        #Making text
        circle = Circle(color = PINK)
        first_line = TextMobject("Manim is fun")
        second_line = TextMobject("and useful")
        final_line = TextMobject("Hope you like it too!", color=BLUE)
        color_final_line = TextMobject("Hope you like it too!")

        #Coloring
        color_final_line.set_color_by_gradient(BLUE,GREEN)

        #Position text
        second_line.next_to(first_line, DOWN)

        #Showing text
        self.wait(1)
        self.play(Write(first_line), Write(second_line))
        self.wait(1)
        self.play(FadeOut(second_line), ReplacementTransform(first_line, circle))
        self.wait(1)
        self.play(Transform(final_line, color_final_line))
        self.wait(2)
```

```python
%%manim Shapes -l

class Shapes(Scene):
    def construct(self):
        #######Code#######
        #Making Shapes
        circle = Circle(color=GREEN)
        square = Square(color=DARK_BLUE)
        square.surround(circle)

        rectangle = Rectangle(height=2, width=3, color=RED)
        ring=Annulus(inner_radius=.2, outer_radius=1, color=BLUE)
        ring2 = Annulus(inner_radius=0.6, outer_radius=1, color=BLUE)
        ring3=Annulus(inner_radius=.2, outer_radius=1, color=BLUE)
        ellipse=Ellipse(width=5, height=3, color=DARK_BLUE)

        pointers = []
        for i in range(8):
            pointers.append(Line(ORIGIN, np.array([cos(pi/180*360/8*i),sin(pi/180*360/8*i), 0]),color=YELLOW))

        #Showing animation
        self.add(circle)
        self.play(FadeIn(square))
        self.play(Transform(square, rectangle))
        self.play(FadeOut(circle), FadeIn(ring))
        self.play(Transform(ring,ring2))
        self.play(Transform(ring2, ring))
        self.play(FadeOut(square), GrowFromCenter(ellipse), Transform(ring2, ring3))
        self.add(*pointers)
        self.wait(1)
```

```python
%%manim 

class Grid(VGroup):
    CONFIG = {
        "height": 6.0,
        "width": 6.0,
    }

    def __init__(self, rows, columns, **kwargs):
        digest_config(self, kwargs, locals())
        super().__init__(**kwargs)

        x_step = self.width / self.columns
        y_step = self.height / self.rows

        for x in np.arange(0, self.width + x_step, x_step):
            self.add(Line(
                [x - self.width / 2., -self.height / 2., 0],
                [x - self.width / 2., self.height / 2., 0],
            ))
        for y in np.arange(0, self.height + y_step, y_step):
            self.add(Line(
                [-self.width / 2., y - self.height / 2., 0],
                [self.width / 2., y - self.height / 2., 0]
            ))


class ScreenGrid(VGroup):
    CONFIG = {
        "rows": 8,
        "columns": 14,
        "height": FRAME_Y_RADIUS * 2,
        "width": 14,
        "grid_stroke": 0.5,
        "grid_color": WHITE,
        "axis_color": RED,
        "axis_stroke": 2,
        "labels_scale": 0.25,
        "labels_buff": 0,
        "number_decimals": 2
    }

    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        rows = self.rows
        columns = self.columns
        grid = Grid(width=self.width, height=self.height, rows=rows, columns=columns)
        grid.set_stroke(self.grid_color, self.grid_stroke)

        vector_ii = ORIGIN + np.array((- self.width / 2, - self.height / 2, 0))
        vector_si = ORIGIN + np.array((- self.width / 2, self.height / 2, 0))
        vector_sd = ORIGIN + np.array((self.width / 2, self.height / 2, 0))



        divisions_x = self.width / columns
        divisions_y = self.height / rows

        directions_buff_x = [UP, DOWN]
        directions_buff_y = [RIGHT, LEFT]
        dd_buff = [directions_buff_x, directions_buff_y]
        vectors_init_x = [vector_ii, vector_si]
        vectors_init_y = [vector_si, vector_sd]
        vectors_init = [vectors_init_x, vectors_init_y]
        divisions = [divisions_x, divisions_y]
        orientations = [RIGHT, DOWN]
        labels = VGroup()
        set_changes = zip([columns, rows], divisions, orientations, [0, 1], vectors_init, dd_buff)
        for c_and_r, division, orientation, coord, vi_c, d_buff in set_changes:
            for i in range(1, c_and_r):
                for v_i, directions_buff in zip(vi_c, d_buff):
                    ubication = v_i + orientation * division * i
                    coord_point = round(ubication[coord], self.number_decimals)
                    label = Text(f"{coord_point}",font="Arial",stroke_width=0).scale(self.labels_scale)
                    label.next_to(ubication, directions_buff, buff=self.labels_buff)
                    labels.add(label)

        self.add(grid, labels)


class CoordScreen(Scene):
    
    def construct(self):
        screen_grid = ScreenGrid()
      
        self.add(screen_grid)
        CoordScreen.make_list(self, [3,2,1,5,4,8])

    
    
    def make_list(self, ray):
        
        ### the following code set up the array
        left_orgin = np.array([-4.5,2.5, 0])
        array = []
        val = []
        for i in range(len(ray)):
            One_Square = Square(side_length = 1)
            One_Square.move_to(left_orgin + i * RIGHT)
            One_val = TextMobject(str(ray[i]))
            One_val.move_to(One_Square.get_center())
            array.append(One_Square)
            val.append(One_val)
        
        V_array = VGroup(*array)
        V_val = VGroup(*val)
        self.play(ShowCreation(V_array, run_time=2))
        self.play(ShowCreation(V_val))
        
        
        ### the following code set up pre-iteration animati9on
        code1 = TextMobject( "for i in intlist :"  )
        code1.shift(UP*0.5)
        code1.shift(LEFT*4)
        code2 = TextMobject( " sum = sum + i")
        code2.next_to(code1, DOWN)
        code2.shift(RIGHT)
        V_code_1 = VGroup(code1, code2)
        V_code_1.set_color(YELLOW)
        
        i = "?"
        sum_value = 0
        
        line_1 = TextMobject(" i = ")
        line_1.shift(LEFT*4.7)
        line_2 = TextMobject("sum = ")
        line_2.next_to(line_1, DOWN)
        i_val = TextMobject(i)
        sum_val = TextMobject(str(sum_value))
        
        i_val.next_to(line_1, RIGHT)
        sum_val.next_to(line_2, RIGHT)
        
        code_sq = Rectangle(color = YELLOW)
        
        V_code_2 = VGroup(line_1, line_2, i_val, sum_val)
        code_sq.surround(V_code_2)
        
        arrow = Arrow([0,0,0],[0,2,0])
        arrow.next_to(V_code_2, UP)
        
        V_block = VGroup(V_code_2, code_sq, arrow)
        
        self.play(Write(V_code_1))
        self.wait(1)
        
        self.play(ReplacementTransform(V_code_1, V_block))
        
        V_block.generate_target()
        V_block.target.move_to(V_val[0].get_center() + DOWN *2)
  
        self.play(MoveToTarget(V_block))
    
    
    
        self.wait(1)
        
        
        ### add the blinking message next to the V_block
        code_3 = TextMobject("sum = sum + i")
        code_3.scale(0.9)
        code_3.set_color(PINK)


  
        
        ### the follow lines of code starts iterating
        for i in range(len(ray)):
            self.play(FadeOut(i_val))
            
            current_i = val[i].copy()
            current_i.set_color(GREEN_SCREEN)
            
            
            #shifting number from list into box
            current_i.generate_target()
            current_i.target.move_to(i_val.get_center())
            self.play(MoveToTarget(current_i))
            i_val.become(current_i)
            
            
            #blink the line of code3
            code_3.next_to(sum_val, RIGHT*2)
            
            #Update the value of Sum
            sum_value = sum_value + ray[i]
            new_sum_val = TextMobject(str(sum_value))
            new_sum_val.set_color(PINK)
            new_sum_val.move_to(sum_val)
            self.play(FadeIn(code_3))
            self.play(ReplacementTransform(sum_val, new_sum_val, runtime = 1), FadeOut(code_3, runtime = 2))
            
            
            
            #shifting box
            self.remove(new_sum_val)
            self.remove(current_i)
            V_block.generate_target()
            V_block.target.shift(RIGHT * 1)
            self.play(MoveToTarget(V_block))


            
            
            
            
            
            
        
        
        
        
        
        
            
            

            
            
        
        
        
        
        
```
