```python
%%manim 


class CoordScreen(Scene):
    
    def construct(self):
        #screen_grid = ScreenGrid()
      
        #self.add(screen_grid)
        CoordScreen.make_list(self, [3,2,1,5,4,8, 5, 7])

    
    
    def make_list(self, ray):
        
        ### the following code set up the array
        left_orgin = np.array([-4.5,2.5, 0])
        array = []
        val = []
        
        #ray is a python list containing actual value
        #array is the list containing the LinkList Block Object
        #val is the list containing the str representation of the values inside the linked list
        
        for i in range(len(ray)):
            One_Square = Square(side_length = 1)
            One_Square.move_to(left_orgin + i * RIGHT)
            One_val = TextMobject(str(ray[i]))
            One_val.move_to(One_Square.get_center())
            array.append(One_Square)
            val.append(One_val)
        
        V_array = VGroup(*array)
        V_val = VGroup(*val)
        self.play(ShowCreation(V_array, run_time=2))
        self.play(ShowCreation(V_val))
        
        
        ### the following code set up pre-iteration animati9on
        code1 = TextMobject( "for ", "i "  , " in ", " intlist :"  )
        code1[0].set_color(GREEN_SCREEN)
        code1[2].set_color(GREEN_SCREEN)
        code1.shift(UP*0.5)
        code1.shift(LEFT*4)
        code2 = TextMobject( " total ", "= ", "total " ,"+ i")
        code2[0].set_color(PINK)
        code2[2].set_color(PINK)
        code2.next_to(code1, DOWN)
        code2.shift(RIGHT)
        V_code_1 = VGroup(code1, code2)
        
        i = "?"
        sum_value = 0
        
        line_1 = TextMobject(" i = ")
        line_1.shift(LEFT*4.7)
        line_2 = TextMobject("total = ")
        line_2.next_to(line_1, DOWN)
        i_val = TextMobject(i)
        sum_val = TextMobject(str(sum_value))
        
        i_val.next_to(line_1, RIGHT)
        sum_val.next_to(line_2, RIGHT)
        
        code_sq = Rectangle(color = YELLOW)
        
        V_code_2 = VGroup(line_1, line_2, i_val, sum_val)
        code_sq.surround(V_code_2)
        
        arrow = Arrow([0,0,0],[0,2,0])
        arrow.next_to(V_code_2, UP)
        
        V_block = VGroup(V_code_2, code_sq, arrow)
        
        self.play(Write(V_code_1))
        self.wait(1)
        
        self.play(ReplacementTransform(V_code_1, V_block))
        
        V_block.generate_target()
        V_block.target.move_to(V_val[0].get_center() + DOWN *2)
  
        self.play(MoveToTarget(V_block))
    
    
    
        self.wait(1)
        
        
        ### add the blinking message next to the V_block
        code_3 = TextMobject("total = total + i")
        code_3.scale(0.9)
        code_3.set_color(PINK)


  
        
        ### the follow lines of code starts iterating
        for i in range(len(ray)):
            self.play(FadeOut(i_val))
            
            current_i = val[i].copy()
            current_i.set_color(GREEN_SCREEN)
            
            
            #shifting number from list into box
            current_i.generate_target()
            current_i.target.move_to(i_val.get_center())
            self.play(MoveToTarget(current_i))
            i_val.become(current_i)
            
            
            #blink the line of code3
            code_3.next_to(sum_val, RIGHT*2)
            
            #Update the value of Sum
            sum_value = sum_value + ray[i]
            new_sum_val = TextMobject(str(sum_value))
            new_sum_val.set_color(PINK)
            new_sum_val.move_to(sum_val)
            self.play(FadeIn(code_3))
            self.play(ReplacementTransform(sum_val, new_sum_val, runtime = 1), FadeOut(code_3, runtime = 2))
            
            
            
            #shifting box
            self.remove(new_sum_val)
            self.remove(current_i)
            V_block.generate_target()
            V_block.target.shift(RIGHT * 1)
            self.play(MoveToTarget(V_block))


            
            
            
            
            
            
        
        
        
        
        
        
            
            

            
            
        
        
        
        
        
```
